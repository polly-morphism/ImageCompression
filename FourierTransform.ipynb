```python
from numpy import mean, zeros, size, flipud, real, transpose, array, log, shape, absolute, e, pi, exp, vectorize
import cv2 as cv
from functools import reduce
from matplotlib import pyplot as plt

img = cv.imread('64by64.jpg') 
test = mean(img, 2)

def fourier_transform(image_matrix):
    N = int(shape(image_matrix)[0])
    new_matrix = zeros(shape(image_matrix), dtype=complex)
    for k in range(N-1):
        for l in range(N-1):
            F_k_l = 0
            for n in range(N-1):
                for m in range(N-1):
                    F_k_l += complex(image_matrix[n,m])*exp(-1*1j*2*pi*((k*m)/N + (l*m)/N))
            new_matrix[n,m] = F_k_l
    return(new_matrix)

            
new_matrix = fourier_transform(test)


plt.style.use('grayscale')
print("Original")
plt.imshow(test)
plt.show()
plt.imshow(absolute(new_matrix))
plt.show()
```
